## 步骤2[可选]、将图片调整为800*800尺寸

```python
import os
import io
import tensorflow as tf
from PIL import Image
import pandas as pd
from collections import namedtuple, OrderedDict
from tqdm import tqdm
import random
import time

FullIJCNN2013_label_path = "/home/wilson/workstation/my_traffic_tf2/data/gtsdb.label"
FullIJCNN2013_path = "/home/wilson/workstation/my_traffic_tf2/data/FullIJCNN2013"
PPM_IMG_dir = FullIJCNN2013_path

FullIJCNN2013_800x800_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013_800x800"

YOLO3_dataset_train_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-train.dataset"
YOLO3_dataset_test_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-test.dataset"
YOLO3_dataset_val_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-val.dataset"

MUST_WIDTH = 1360
MUST_HEIGHT = 800
TARGET_SIZE = 800
Const_Image_Format = [".ppm"]

#######################################################
# 检测每张图片的尺寸
def check_image_size():
    print("checking ppm image size must: (%d,%d)..." %(MUST_WIDTH, MUST_HEIGHT))
    for name in os.listdir(PPM_IMG_dir):
        pathName = os.path.join(PPM_IMG_dir, name)
        if os.path.isfile(pathName):
            if os.path.splitext(pathName)[1] in Const_Image_Format:
                with tf.io.gfile.GFile(pathName, "rb") as fid:
                    encoded_jpg = fid.read()
                    encoded_jpg_io = io.BytesIO(encoded_jpg)
                    image = Image.open(encoded_jpg_io)
                    originalWidth = image.size[0]
                    originalHeight = image.size[1]
                    if originalWidth != MUST_WIDTH or originalHeight != MUST_HEIGHT:
                          print(name, "%s size invalid: width:%d, height:%d" %(pathName, originalWidth, originalHeight))
    print("checking ppm image done.")

#########################################################
def split(df, group):
    data = namedtuple('data', ['img', 'object'])
    gb = df.groupby(group)
    return [data(img, gb.get_group(x)) for img, x in zip(gb.groups.keys(), gb.groups)]

##########################################################
# 开始将图片转成 800*800的规格，并且包含分类的框框信息
# 将原始分类信息按照所在图片进行整理

#check_image_size()

data = pd.read_csv(FullIJCNN2013_path+ "/gt.txt", sep=";", header=None)
data.columns = ["img", "x1", "y1", "x2", "y2", "id"]

top3 = data[:3]
print("the top 3:\n", top3)

# 框框列表
boxes = []

groupedByImg = split(data, "img")
for group in tqdm(groupedByImg):
    # 获取所有框框的外切矩形
    X1, Y1, X2, Y2 = None, 0, None, TARGET_SIZE-1
    for index, row in group.object.iterrows():
        # Y 轴不处理，固定为800
        x1, x2 = int(row["x1"]), int(row["x2"])
        if X1 == None or x1 < X1: X1 = x1
        if X2 == None or x2 > X2: X2 = x2
        
    # 由框框组成的长度
    valid_width = X2 - X1
    
    # 新图片的存储路径
    savePath = os.path.join(FullIJCNN2013_800x800_path, group.img[:-3] + 'jpg')
    
    if valid_width >= TARGET_SIZE:
        # 最左边一部分内容会被截掉
        cutX = X1
        for index, row in group.object.iterrows():
            # 每个x轴砍掉 cutX 长度
            x1, y1, x2, y2, id = int(row["x1"]) - cutX, int(row["y1"]), int(row["x2"]) - cutX, int(row["y2"]), int(row["id"])
            boxes.append([savePath, x1, y1, x2, y2, id])
        # 进行图像截取
        with tf.io.gfile.GFile(os.path.join(FullIJCNN2013_path, group.img), "rb") as fid:
            encoded_jpg = fid.read()
            encoded_jpg_io = io.BytesIO(encoded_jpg)
            image = Image.open(encoded_jpg_io)
            image = image.crop((X1, Y1, X2, Y2))
            image = image.resize((TARGET_SIZE, TARGET_SIZE),Image.BILINEAR)        
            image.save(savePath)
        
    else: #valid_width < TARGET_SIZE 说明框框很小，需要进行扩充
        total_need_width = TARGET_SIZE - valid_width
        #  优先往左边扩充，尽量让交通标记在右边，像正常实际的情况
        left_width = X1
        if left_width >= total_need_width: # 左边空间足够
            # 最左边一部分内容会被截掉
            cutX = left_width - total_need_width
            for index, row in group.object.iterrows():
                # 每个x轴砍掉 cutX 长度
                x1, y1, x2, y2, id = int(row["x1"]) - cutX, int(row["y1"]), int(row["x2"]) - cutX, int(row["y2"]), int(row["id"])
                boxes.append([savePath, x1, y1, x2, y2, id])
            # 进行图像截取
            with tf.io.gfile.GFile(os.path.join(FullIJCNN2013_path, group.img), "rb") as fid:
                encoded_jpg = fid.read()
                encoded_jpg_io = io.BytesIO(encoded_jpg)
                image = Image.open(encoded_jpg_io)
                image = image.crop((cutX, Y1, X2, Y2))
                image.save(savePath)
                
        else:# 左边部分不足，直接从最坐标开始截取即可，这样整个外切矩形会处于图片中间
            for index, row in group.object.iterrows():
                # 每个x轴保持即可
                x1, y1, x2, y2, id = int(row["x1"]), int(row["y1"]), int(row["x2"]), int(row["y2"]), int(row["id"])
                boxes.append([savePath, x1, y1, x2, y2, id])
            # 进行图像截取
            with tf.io.gfile.GFile(os.path.join(FullIJCNN2013_path, group.img), "rb") as fid:
                encoded_jpg = fid.read()
                encoded_jpg_io = io.BytesIO(encoded_jpg)
                image = Image.open(encoded_jpg_io)
                image = image.crop((0, Y1, TARGET_SIZE, Y2))
                image.save(savePath)

# 打乱框框列表
r=random.random
random.seed(int(time.time()))
random.shuffle(boxes, random=r)

# 获取三个集合
train_ratio = 0.7
test_ratio = 0.5
train_count = int(len(boxes) * train_ratio)
remain = len(boxes) - train_count
test_count = int(remain * test_ratio)
val_count = remain - test_count
print("\ntrain count:", train_count)
print(  " test count:", test_count)
print(  " val  count:", val_count)

xtrain = boxes[0:train_count]
xtest = boxes[train_count:train_count+test_count]
xval = boxes[train_count+test_count:]

# 对每个集合里的内容，按照图片进行组合
def imageMapBoxes(bs, dataset_save_path):
    with open(dataset_save_path, "w", encoding="utf-8") as file:   #”w"代表着每次运行都覆盖内容
        # 先清空内容
        file.seek(0)
        file.truncate()
        
        imageMap = {}
        for box in bs:
            img, x1, y1, x2, y2, id = box[0], box[1], box[2], box[3], box[4], box[5]
            if img in imageMap:
                list = imageMap[img]
                list.append([x1, y1, x2, y2, id])
            else:
                list = []
                imageMap[img] = list
                list.append([x1, y1, x2, y2, id])
        
        for img, list in imageMap.items():
            array = []
            array.append(img)
            for b in list:
                array.append("%d,%d,%d,%d,%d" % (b[0], b[1], b[2], b[3], b[4]))
            file.write(" ".join(array) + os.linesep)
        
imageMapBoxes(xtrain, YOLO3_dataset_train_path)
imageMapBoxes(xtest, YOLO3_dataset_test_path)
imageMapBoxes(xval, YOLO3_dataset_val_path)

print("Done")
```

## 将图片绘制框框

```python
import fileinput 
import re
import os
import cv2
from PIL import Image
from PIL import ImageDraw
from tqdm import tqdm

FullIJCNN2013_800x800_boxes_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013_800x800_boxes"

YOLO3_dataset_train_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-train.dataset"
YOLO3_dataset_test_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-test.dataset"
YOLO3_dataset_val_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-val.dataset"

"""获取文件名称"""
def get_file_name(path_string):
    pattern = re.compile(r'([^<>/\\\|:""\*\?]+)\.(\w+)$')
    data = pattern.findall(path_string)
    if data:
        return data[0][0]+"."+data[0][1]
    
def drawBoxesOnImage(dataset_path):
    file = open(dataset_path)
    for line in tqdm(file):
        elems = line.split(" ")
        # 原始目录
        imagePath, boxes = elems[0], elems[1:]
        # 可能已经存在该图片了
        imageName = get_file_name(imagePath)
        imageSavePath = os.path.join(FullIJCNN2013_800x800_boxes_path, imageName)
        img = None
        if os.path.exists(imageSavePath):
            img = cv2.imread(imageSavePath) #打开图片1.jpg    
        else:
            img = cv2.imread(imagePath) #打开图片1.jpg
        
        for bStr in boxes:
            box = bStr.split(",")
            x1, y1, x2, y2 = int(box[0]), int(box[1]), int(box[2]), int(box[3])
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)

        cv2.imwrite(imageSavePath, img)

drawBoxesOnImage(YOLO3_dataset_train_path)
drawBoxesOnImage(YOLO3_dataset_test_path)
drawBoxesOnImage(YOLO3_dataset_val_path)
```

## 导出yolo的尺寸数据集

```python
import os
import tqdm

FullIJCNN2013_label_path = "/home/wilson/workstation/my_traffic_tf2/data/gtsdb.label"
FullIJCNN2013_800x800_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013_800x800"

source_dataset_train_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-train.dataset"
source_dataset_test_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-test.dataset"
source_dataset_val_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013-val.dataset"

YOLO3_dataset_train_path = "./data/dataset/FullIJCNN2013-train.dataset"
YOLO3_dataset_test_path = "./data/dataset/FullIJCNN2013-test.dataset"
YOLO3_dataset_val_path = "./data/dataset/FullIJCNN2013-val.dataset"

YOLO3_jpg_dir = "./data/jpgFullIJCNN2013"
YOLO3_classnames_path = "./data/classes/FullIJCNN2013.names"

TARGET_SIZE = 608
SIZE_SCALE = 608 * 1.0 / 800
Const_Image_Format = [".jpg"]

#####################################
# 转存类别名称
def save_labels():
    with open(YOLO3_classnames_path, 'w', encoding="utf-8") as fwriter:
        with open(FullIJCNN2013_label_path, 'r', encoding="utf-8") as freader:
            lines = freader.readlines()
            for line in lines:
                if not line.split():
                    continue
                line = line.strip()
                _, name = line.split(' ', 1)
                fwriter.write(name+"\n")
                
#########################################
def process(source_path, target_path):
    with open(target_path, 'w', encoding="utf-8") as fwriter:
        with open(source_path, 'r', encoding="utf-8") as freader:
            lines = freader.readlines()
            for line in lines:
                if not line.split():
                    continue
                line = line.strip()
                elems = line.split(" ")
                imgName = elems[0]
                boxes = elems[1:]
                array = []
                array.append(imgName)
                for bStr in boxes:
                    box = bStr.split(",")
                    x1, y1, x2, y2, id = int(box[0]), int(box[1]),int(box[2]),int(box[3]),int(box[4])
                    x1 = int(x1 * SIZE_SCALE)
                    y1 = int(y1 * SIZE_SCALE)
                    x2 = int(x2 * SIZE_SCALE)
                    y2 = int(y2 * SIZE_SCALE)
                    array.append("%d,%d,%d,%d,%d" %(x1, y1, x2, y2, id))
                fwriter.write(" ".join(array)+"\n")

# 转换输入的图片集合
for name in os.listdir(FullIJCNN2013_800x800_path):
    pathName = os.path.join(FullIJCNN2013_800x800_path, name)
    if os.path.isfile(pathName):
        if os.path.splitext(pathName)[1] in Const_Image_Format:
            with tf.io.gfile.GFile(pathName, "rb") as fid:
                encoded_jpg = fid.read()
                encoded_jpg_io = io.BytesIO(encoded_jpg)
                image = Image.open(encoded_jpg_io)
                image = image.resize((TARGET_SIZE, TARGET_SIZE),Image.BILINEAR)
                image.save(os.path.join(YOLO3_jpg_dir, name))

# 转变数据集尺寸
process(source_dataset_train_path, YOLO3_dataset_train_path)
process(source_dataset_test_path, YOLO3_dataset_test_path)
process(source_dataset_val_path, YOLO3_dataset_val_path)

print("Done")
```

## 步骤2、导出yolo使用的数据集

```python
""" 
FullIJCNN2013 的数据集格式 gt.txt:
img;x1;y1;x2;y2;id

--------

yolo3 的数据集所需格式 FullIJCNN2013.dataset:
xxx/xxx.jpg 18.19,6.32,424.13,421.83,20 323.86,2.65,640.0,421.94,20 
xxx/xxx.jpg 48,240,195,371,11 8,12,352,498,14
image_path x_min,y_min,x_max,y_max,class_id  x_min,y_min,...,class_id 
(注意空格和逗号)
"""

# 需要预处理三大部分内容：
# 1、训练所需要的图片集合：YOLO3_jpg_dir
# 2、每张图片对应的 类别和框框列表：YOLO3_dataset_path
# 3、类别映射的名称列表

import os
import io
import time
import pandas as pd
import numpy as np
import tensorflow as tf
from tqdm import tqdm
from PIL import Image
from collections import namedtuple, OrderedDict
from sklearn.model_selection import KFold
import matplotlib.pyplot as plt

%matplotlib inline

# 分三种224*224,416*416，608*608三种。图片越大，时间越长，但是并不是resize耗费的时间比较长，是卷积的次数多。
TARGET_SIZE = 608
VAL_SIZE = 608

# 导出的数据集格式输出位置
YOLO3_dataset_train_path = "./data/dataset/FullIJCNN2013-train.dataset"
YOLO3_dataset_test_path = "./data/dataset/FullIJCNN2013-test.dataset"
YOLO3_dataset_val_path = "./data/dataset/FullIJCNN2013-val.dataset"
YOLO3_jpg_dir = "./data/jpgFullIJCNN2013"
YOLO3_classnames_path = "./data/classes/FullIJCNN2013.names"

FullIJCNN2013_label_path = "/home/wilson/workstation/my_traffic_tf2/data/gtsdb.label"
FullIJCNN2013_path = "/home/wilson/workstation/my_traffic_tf2/data/FullIJCNN2013"
PPM_IMG_dir = FullIJCNN2013_path

# 数据集输入的jpg图片所在目录
if not os.path.exists(YOLO3_jpg_dir):
    os.mkdir(YOLO3_jpg_dir)

def split(df, group):
    data = namedtuple('data', ['img', 'object'])
    gb = df.groupby(group)
    return [data(img, gb.get_group(x)) for img, x in zip(gb.groups.keys(), gb.groups)]

#####################################
# 转换输入的图片集合
def save_ppm_as_jpg(ppmFilename, jpgFilepath, target_size):
    ppm_img_path = os.path.join(PPM_IMG_dir, ppmFilename)
    with tf.io.gfile.GFile(ppm_img_path, "rb") as fid:
        encoded_jpg = fid.read()
        encoded_jpg_io = io.BytesIO(encoded_jpg)
        image = Image.open(encoded_jpg_io)
        image = image.resize((target_size, target_size),Image.BILINEAR)
        image.save(jpgFilepath)

########################################
# 存储 每张图片对应的 类别和框框列表
def save_img_dataset(group, jpgFilepath, dataset_path, target_size):   
    ppm_img_path = os.path.join(PPM_IMG_dir, group.img)
    with tf.io.gfile.GFile(ppm_img_path, "rb") as fid:
        encoded_jpg = fid.read()
        encoded_jpg_io = io.BytesIO(encoded_jpg)
        image = Image.open(encoded_jpg_io)
        originalWidth = image.size[0]
        originalHeight = image.size[1]
        widthScale = target_size * 1.0 / originalWidth
        heightScale = target_size * 1.0 / originalHeight
        
    with open(dataset_path,"a") as file:   #”w"代表着每次运行都覆盖内容
        groupIds = []
        groupIds.append(jpgFilepath)
        for index, row in group.object.iterrows():
            boxIdBlock = []
            boxIdBlock.append(str(int(row["x1"] * widthScale)))
            boxIdBlock.append(str(int(row["y1"] * heightScale)))
            boxIdBlock.append(str(int(row["x2"] * widthScale)))
            boxIdBlock.append(str(int(row["y2"] * heightScale)))
            boxIdBlock.append(str(int(row["id"])))
            groupIds.append(",".join(boxIdBlock))
        line = " ".join(groupIds)+"\n"
        file.write(line)

######################################
# 清空数据集文件
def clearDatasetFile():
    for dataset_path in [YOLO3_dataset_train_path, YOLO3_dataset_test_path, YOLO3_dataset_val_path]:
        with open(dataset_path, "w") as file:   #”w"代表着每次运行都覆盖内容
            file.seek(0)
            file.truncate()

# 转存类别名称
def save_labels():
    with open(YOLO3_classnames_path, 'w', encoding="utf-8") as fwriter:
        with open(FullIJCNN2013_label_path, 'r', encoding="utf-8") as freader:
            lines = freader.readlines()
            for line in lines:
                if not line.split():
                    continue
                line = line.strip()
                _, name = line.split(' ', 1)
                fwriter.write(name+"\n")
                
###################################
# 按照图片名进行进行打组
def process(df, dataset_path, target_size):
    groupedByImg = split(df, "img")
    for group in tqdm(groupedByImg):
        jpgFilename = group.img[:-3] + 'jpg'
        jpgFilepath = os.path.join(YOLO3_jpg_dir, jpgFilename)
        
        # 将原始的ppm图片转为 jpg格式并存档
        save_ppm_as_jpg(group.img, jpgFilepath, target_size)
        
        # 存储 每张图片对应的 类别和框框列表
        save_img_dataset(group, jpgFilepath, dataset_path, target_size)

#######################################
clearDatasetFile()
save_labels()

data = pd.read_csv(FullIJCNN2013_path+ "/gt.txt", sep=";", header=None)
data.columns = ["img", "x1", "y1", "x2", "y2", "id"]

top3 = data[:3]
print("the top 3:\n", top3)

# 将分类只有一条记录的数据，拼接到尾部
threshold = 1
y_less = data.groupby("id").filter(lambda x: len(x) <= threshold)
data = pd.concat([data, y_less], ignore_index=True)

# 类型对应所有数量
idValueList = data['id']
#print('\nid value list:\n', idValueList)

random_state = int(time.time())
kf = KFold(n_splits=2, shuffle=True, random_state=random_state)
#print('\nthe kf is:\n', kf)

shuffles = np.hstack((next(kf.split(idValueList))))
print("\nshuffles count:", len(shuffles))

train_ratio = 0.7
test_ratio = 0.5
train_count = int(idValueList.count() * train_ratio)
remain = idValueList.count() - train_count
test_count = int(remain * test_ratio)
val_count = remain - test_count
print("\ntrain count:", train_count)
print(  " test count:", test_count)
print(  " val  count:", val_count)

xtrain = data.iloc[0:train_count,:]
xtest = data.iloc[train_count:train_count+test_count,:]
xval = data.iloc[train_count+test_count:,:]

plt.hist([
        xtrain['id'], 
        xtest['id'], 
        xval['id'],
        ], 
        stacked=True, 
        label=["train", "test", "val"],
        bins=86
    )
plt.show()

process(xtrain, YOLO3_dataset_train_path, TARGET_SIZE)
process(xtest, YOLO3_dataset_test_path, TARGET_SIZE)
process(xval, YOLO3_dataset_val_path, VAL_SIZE)

print("Done.")
```
