## 步骤2、将德国交通标志图片调整为800*800尺寸

```python
import os
import io
import tensorflow as tf
from PIL import Image
import pandas as pd
from collections import namedtuple, OrderedDict
from tqdm import tqdm
import random
import time
import matplotlib.pyplot as plt

%matplotlib inline

FullIJCNN2013_path = "/home/wilson/workstation/my_traffic_tf2/data/FullIJCNN2013"
PPM_IMG_dir = FullIJCNN2013_path

FullIJCNN2013_800x800_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013_800x800"

MUST_WIDTH = 1360
MUST_HEIGHT = 800
TARGET_SIZE = 800
Const_Image_Format = [".ppm"]

#######################################################
# 检测每张图片的尺寸
def check_image_size():
    print("checking ppm image size must: (%d,%d)..." %(MUST_WIDTH, MUST_HEIGHT))
    for name in os.listdir(PPM_IMG_dir):
        pathName = os.path.join(PPM_IMG_dir, name)
        if os.path.isfile(pathName):
            if os.path.splitext(pathName)[1] in Const_Image_Format:
                with tf.io.gfile.GFile(pathName, "rb") as fid:
                    encoded_jpg = fid.read()
                    encoded_jpg_io = io.BytesIO(encoded_jpg)
                    image = Image.open(encoded_jpg_io)
                    originalWidth = image.size[0]
                    originalHeight = image.size[1]
                    if originalWidth != MUST_WIDTH or originalHeight != MUST_HEIGHT:
                          print(name, "%s size invalid: width:%d, height:%d" %(pathName, originalWidth, originalHeight))
    print("checking ppm image done.")

#########################################################
def split(df, group):
    data = namedtuple('data', ['img', 'object'])
    gb = df.groupby(group)
    return [data(img, gb.get_group(x)) for img, x in zip(gb.groups.keys(), gb.groups)]

##########################################################
# 开始将图片转成 800*800的规格，并且包含分类的框框信息
# 将原始分类信息按照所在图片进行整理

#check_image_size()

data = pd.read_csv(FullIJCNN2013_path+ "/gt.txt", sep=";", header=None)
data.columns = ["img", "x1", "y1", "x2", "y2", "id"]

top3 = data[:3]
print("the top 3:\n", top3)

# 框框列表
# elem: [imagePath,x1,y1,x2,y2,id]
boxes = []

groupedByImg = split(data, "img")
for group in tqdm(groupedByImg):
    # 获取所有框框的外切矩形
    X1, Y1, X2, Y2 = None, 0, None, TARGET_SIZE-1
    for index, row in group.object.iterrows():
        # Y 轴不处理，固定为800
        x1, x2 = int(row["x1"]), int(row["x2"])
        if X1 == None or x1 < X1: X1 = x1
        if X2 == None or x2 > X2: X2 = x2
        
    # 由框框组成的长度
    valid_width = X2 - X1
    
    # 扩宽10个像素，避免框框贴到边缘，造成计算问题。
    padding = 10
    valid_width = valid_width + 2 * padding
    
    # 新图片的存储路径
    savePath = os.path.join(FullIJCNN2013_800x800_path, group.img[:-3] + 'jpg')
        
    if valid_width >= TARGET_SIZE:
        # 最左边一部分内容会被截掉
        left = X1 - padding
        right = X2 + padding
        if left < 0:
            right += math.abs(left)
            left = 0
        
        if right >= MUST_WIDTH:
            delta = 1 + (right - MUST_WIDTH)
            right = right - delta
            left = left - delta
        
        # 需要进行缩放
        width_scale = TARGET_SIZE * 1.0 / (right - left)
            
        left_padding = X1 - left
        for index, row in group.object.iterrows():
            # 每个x轴砍掉 cutX 长度
            x1 = int((int(row["x1"]) - X1 + left_padding) * width_scale)
            y1 = int(row["y1"])
            x2 = int((int(row["x2"]) - X1 + left_padding) * width_scale)
            y2 = int(row["y2"])
            id = int(row["id"])
            boxes.append([savePath, x1, y1, x2, y2, id])
        
        # 进行图像截取
        with tf.io.gfile.GFile(os.path.join(FullIJCNN2013_path, group.img), "rb") as fid:
            encoded_jpg = fid.read()
            encoded_jpg_io = io.BytesIO(encoded_jpg)
            image = Image.open(encoded_jpg_io)
            image = image.crop((left, Y1, right, Y2))
            image = image.resize((TARGET_SIZE, TARGET_SIZE),Image.BILINEAR)        
            image.save(savePath)
        
    else: #valid_width < TARGET_SIZE 说明框框很小，需要进行扩充
        total_need_width = TARGET_SIZE - valid_width
        #  优先往左边扩充，尽量让交通标记在右边，像正常实际的情况
        left_width = X1
        
        if left_width >= total_need_width: # 左边空间足够
            # 最左边一部分内容会被截掉
            left = left_width - total_need_width + padding
            right = left + TARGET_SIZE
            if right >= MUST_WIDTH:
                delta = 1 + (right - MUST_WIDTH)
                right = right - delta
                left = left - delta
            
            left_padding = X1 - left
            for index, row in group.object.iterrows():
                # 每个x轴砍掉 cutX 长度
                x1 = int(row["x1"]) - X1 + left_padding
                y1 = int(row["y1"])
                x2 = int(row["x2"]) - X1 + left_padding
                y2 = int(row["y2"])
                id = int(row["id"])
                boxes.append([savePath, x1, y1, x2, y2, id])
                    
            # 进行图像截取
            with tf.io.gfile.GFile(os.path.join(FullIJCNN2013_path, group.img), "rb") as fid:
                encoded_jpg = fid.read()
                encoded_jpg_io = io.BytesIO(encoded_jpg)
                image = Image.open(encoded_jpg_io)
                image = image.crop((left, Y1, right, Y2))
                image.save(savePath)
                
        else:# 左边部分不足，直接从最坐标开始截取即可，这样整个外切矩形会处于图片中间
            left = X1 - padding
            right = left + TARGET_SIZE
            if left < 0:
                right += math.abs(left)
                left = 0
            if right >= MUST_WIDTH:
                delta = 1 + (right - MUST_WIDTH)
                right = right - delta
                left = left - delta
            
            left_padding = X1 - left
            for index, row in group.object.iterrows():
                x1 = int(row["x1"]) - X1 + left_padding
                y1 = int(row["y1"])
                x2 = int(row["x2"]) - X1 + left_padding
                y2 = int(row["y2"])
                id = int(row["id"])
                boxes.append([savePath, x1, y1, x2, y2, id])
            # 进行图像截取
            with tf.io.gfile.GFile(os.path.join(FullIJCNN2013_path, group.img), "rb") as fid:
                encoded_jpg = fid.read()
                encoded_jpg_io = io.BytesIO(encoded_jpg)
                image = Image.open(encoded_jpg_io)
                image = image.crop((left, Y1, right, Y2))
                image.save(savePath)

print("Done")
```

## 步骤3、处理中国标签，生成yolo3格式

```python
import os
import re
from tqdm import tqdm

""" 中国交通标记 标签所在目录
# chn_数字.txt 格式的文件, 格式：
    <从0开始的类别> <距离左上角的距离x> <距离左上角的距离y> <宽度> <高度>
    <从0开始的类别> <距离左上角的距离x> <距离左上角的距离y> <宽度> <高度>
    ...
"""
ChineseLabelsDir = "/home/wilson/workstation/my_traffic_tf2/data_800x800_chn"
pattern = re.compile(r'chn_(\d+)\.txt$')

def get_image_dir_name(num):
    if num >= 1000 and num <= 1999:
        return "01000~01999"

# 读取标签列表
def read_image_boxes(image_path, label_file_path):
    # 获取图片信息
    with tf.io.gfile.GFile(image_path, "rb") as fid:
        encoded_jpg = fid.read()
        encoded_jpg_io = io.BytesIO(encoded_jpg)
        image = Image.open(encoded_jpg_io)
        image_width = image.size[0]
        image_height = image.size[1]
    
    # 标签信息
    with open(label_file_path, 'r', encoding="utf-8") as freader:
        img_boxes = []
        lines = freader.readlines()
        for line in lines:
            if not line.split():
                continue
            line = line.strip()
            elems = line.split(' ')
            id = int(elems[0])
            box_width = int(image_width * float(elems[3]))
            box_height = int(image_height * float(elems[4]))
            x1 = int(image_width * float(elems[1])) - box_width/2
            y1 = int(image_height * float(elems[2])) - box_height/2
            x2 = x1 + box_width
            y2 = y1 + box_height
            img_boxes.append([image_path, x1, y1, x2, y2, id])
        return img_boxes
                    
# 遍历所有标签文件
for name in tqdm(os.listdir(ChineseLabelsDir)):
    pathName = os.path.join(ChineseLabelsDir, name)
    if os.path.isfile(pathName): # 必须是文件
        found = pattern.findall(name)
        if found:
            img_seqno = int(found[0])
            img_name = name[:-4]+".jpg"
            dir_name = get_image_dir_name(img_seqno)
            image_path = os.path.join(ChineseLabelsDir, dir_name+"/"+img_name)
            
            # 标签列表
            image_boxes = read_image_boxes(image_path, pathName)
            if image_boxes != None and len(image_boxes) > 0:
                boxes.extend(image_boxes)

print("Done. %d boxes generated" % len(boxes))
```

## 步骤4、整合德国/中国 生成train/test/validate 图集

```python
YOLO3_dataset_train_path = "./data/dataset/FullIJCNN2013-train.dataset"
YOLO3_dataset_test_path = "./data/dataset/FullIJCNN2013-test.dataset"
YOLO3_dataset_val_path = "./data/dataset/FullIJCNN2013-val.dataset"

# 打乱框框列表
r=random.random
random.seed(int(time.time()))
random.shuffle(boxes, random=r)

# 获取三个集合
train_ratio = 0.7
test_ratio = 0.5
train_count = int(len(boxes) * train_ratio)
remain = len(boxes) - train_count
test_count = int(remain * test_ratio)
val_count = remain - test_count
print("\ntrain count:", train_count)
print(  " test count:", test_count)
print(  " val  count:", val_count)

xtrain = boxes[0:train_count]
xtest = boxes[train_count:train_count+test_count]
xval = boxes[train_count+test_count:]

# 对每个集合里的内容，按照图片进行组合
def imageMapBoxes(bs, dataset_save_path):
    with open(dataset_save_path, "w", encoding="utf-8") as file:   #”w"代表着每次运行都覆盖内容
        # 先清空内容
        file.seek(0)
        file.truncate()
        
        imageMap = {}
        for box in bs:
            img, x1, y1, x2, y2, id = box[0], box[1], box[2], box[3], box[4], box[5]
            if img in imageMap:
                list = imageMap[img]
                list.append([x1, y1, x2, y2, id])
            else:
                list = []
                imageMap[img] = list
                list.append([x1, y1, x2, y2, id])
        
        for img, list in imageMap.items():
            array = []
            array.append(img)
            for b in list:
                array.append("%d,%d,%d,%d,%d" % (b[0], b[1], b[2], b[3], b[4]))
            file.write(" ".join(array) + os.linesep)
        
imageMapBoxes(xtrain, YOLO3_dataset_train_path)
imageMapBoxes(xtest, YOLO3_dataset_test_path)
imageMapBoxes(xval, YOLO3_dataset_val_path)
print("Done.")
```

## 步骤5、生成anchors

```python
# -*- coding=utf-8 -*-
import glob
import os
import sys
import xml.etree.ElementTree as ET
import numpy as np

sys.path.append("/home/wilson/workstation/kmeans-anchor-boxes")
from kmeans import kmeans, avg_iou

""" 根文件夹（存储训练/测试/校验 的目录，本例子为）
/home/wilson/workstation/tensorflow-yolov3/data/dataset
├── FullIJCNN2013-test.dataset
├── FullIJCNN2013-train.dataset
├── FullIJCNN2013-val.dataset
格式为： <img> x1,y1,x2,y2,id x1,y1,x2,y2,id
"""
ROOT_PATH = '/home/wilson/workstation/tensorflow-yolov3/data/dataset/'
ANCHORS_SAVE_PATH = './data/anchors/my_traffic_anchors.txt'

# 聚类的数目
CLUSTERS = 9

# 模型中图像的输入尺寸，默认是一样的
SIZE = 800

# 加载YOLO格式的标注数据
def load_dataset(dataset, gtfile):
    with open(os.path.join(ROOT_PATH, gtfile), 'r') as f:
        txt_content = f.readlines()
    
    for line in txt_content:
        line = line.strip()
        line_split = line.split(' ')
        for box in line_split[1:]:
            xyids = box.split(',')
            x1, y1, x2, y2 = xyids[0], xyids[1], xyids[2], xyids[3]
            roi_with = float(x2) - float(x1)
            roi_height = float(y2) - float(y1)
            if roi_with == 0 or roi_height == 0:
                continue
            dataset.append([roi_with, roi_height])

data = []
load_dataset(data, "FullIJCNN2013-train.dataset")
load_dataset(data, "FullIJCNN2013-test.dataset")
load_dataset(data, "FullIJCNN2013-val.dataset")
data = np.array(data)

out = kmeans(data, k=CLUSTERS)

print(out)
print("Accuracy: {:.2f}%".format(avg_iou(data, out) * 100))
print("Boxes:\n {}-{}".format(out[:, 0], out[:, 1]))

ratios = np.around(out[:, 0] / out[:, 1], decimals=2).tolist()
print("Ratios:\n {}".format(sorted(ratios)))

with open(ANCHORS_SAVE_PATH, 'w', encoding="utf-8") as fwriter:
    content = []
    for elem in out:
        # 宽高转为 608 等基于32递增的采样大小
        width = float(elem[0])
        height = float(elem[1])
        content.append("%d,%d" % (width, height))
    contentStr = ", ".join(content)
    fwriter.write(contentStr)
    print("anchors written to ", ANCHORS_SAVE_PATH, contentStr)
```

## 步骤6、[可选]将图片绘制框框

```python
import fileinput 
import re
import os
import cv2
from PIL import Image
from PIL import ImageDraw
from tqdm import tqdm
import shutil

FullIJCNN2013_800x800_boxes_path = "/home/wilson/workstation/my_traffic_tf2/data_800x800/FullIJCNN2013_800x800_boxes"

YOLO3_dataset_train_path = "./data/dataset/FullIJCNN2013-train.dataset"
YOLO3_dataset_test_path = "./data/dataset/FullIJCNN2013-test.dataset"
YOLO3_dataset_val_path = "./data/dataset/FullIJCNN2013-val.dataset"

"""获取文件名称"""
def get_file_name(path_string):
    pattern = re.compile(r'([^<>/\\\|:""\*\?]+)\.(\w+)$')
    data = pattern.findall(path_string)
    if data:
        return data[0][0]+"."+data[0][1]
    
def drawBoxesOnImage(dataset_path):
    file = open(dataset_path)
    for line in tqdm(file):
        elems = line.split(" ")
        # 原始目录
        imagePath, boxes = elems[0], elems[1:]
        # 可能已经存在该图片了
        imageName = get_file_name(imagePath)
        imageSavePath = os.path.join(FullIJCNN2013_800x800_boxes_path, imageName)
        img = None
        if os.path.exists(imageSavePath):
            img = cv2.imread(imageSavePath) #打开图片1.jpg    
        else:
            img = cv2.imread(imagePath) #打开图片1.jpg
        
        for bStr in boxes:
            box = bStr.split(",")
            x1, y1, x2, y2 = int(box[0]), int(box[1]), int(box[2]), int(box[3])
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.imwrite(imageSavePath, img)

shutil.rmtree(FullIJCNN2013_800x800_boxes_path)
os.mkdir(FullIJCNN2013_800x800_boxes_path)
drawBoxesOnImage(YOLO3_dataset_train_path)
drawBoxesOnImage(YOLO3_dataset_test_path)
drawBoxesOnImage(YOLO3_dataset_val_path)
print("Done")
```
